## Encoding Features

#### Ordinal Encoding

```python
df.columns
```

```python
# Map blood pressure history to numerical values[Normal: 0, Prehypertension: 1, Hypertension: 2 ]
df['bp_history'] = df['bp_history'].map({'Normal': 0, 'Prehypertension': 1, 'Hypertension': 2}) 

# Map family history to numerical values[No: 0, Yes: 1]
df['family_history'] = df['family_history'].map({'No': 0, 'Yes': 1}) 

# Map exercise level to numerical values[Low: 0, Moderate: 1, High: 2]
df['exercise_level'] = df['exercise_level'].map({'Low': 0, 'Moderate': 1, 'High': 2}) 

# Map smoking status to numerical values[Non-smoker: 0, Smoker: 1]
df['smoking_status'] = df['smoking_status'].map({'Non-Smoker': 0, 'Smoker': 1}) 

# Map hypertension status to numerical values[No: 0, Yes: 1]
df['has_hypertension'] = df['has_hypertension'].map({'No': 0, 'Yes': 1})
```

```python
df
```

#### Nominal Encoding

```python
df = pd.get_dummies(df, columns=['medication'], drop_first=True, dtype=int)
df
```

# Model Training

```python
from  sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X = df.drop('has_hypertension', axis=1)
y = df['has_hypertension']

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=85)
```

```python
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

### Logistic Regression

```python
log_model = LogisticRegression()
log_model.fit(x_train_scaled, y_train)
```

```python
log_model_predictions = log_model.predict(x_test_scaled)
```

```python
print("\nAccuracy:", accuracy_score(y_test, log_model_predictions))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, log_model_predictions))
print("\nClassification Report:\n", classification_report(y_test, log_model_predictions))   
```

### Random Forest

```python
rf_model = RandomForestClassifier(random_state=85)
```

```python
rf_model = RandomForestClassifier(random_state=85)
rf_model.fit(x_train_scaled, y_train)
```

```python
rf_model_predictions = rf_model.predict(x_test_scaled)
```

```python
print("\nAccuracy:", accuracy_score(y_test, rf_model_predictions))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, rf_model_predictions))
print("\nClassification Report:\n", classification_report(y_test, rf_model_predictions))   
```

```python
# Feature Importance 
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

print("\nFeature Importance:\n", feature_importance)
```
